```python
def plot(data, xlabel = "x", ylabel = "y", title = None, saving_name = None, xscale = None, column = 1):
    fig, ax = plt.subplots()
    plt.semilogy(data[0, :,0], data[0, :,column] , "-", color = 'C6', label = '0 deg')
    plt.semilogy(data[1, :,0], data[1, :,column] , "-", color = 'C0', label = '20 deg')
    plt.semilogy(data[2, :,0], data[2, :,column] , "-", color = 'C2', label = '40 deg')

    plt.semilogy(data[3, :,0], data[3, :,column] , "--", color = 'C6', label = '0 deg')
    plt.semilogy(data[4, :,0], data[4, :,column] ,  "--", color = 'C0', label = '20 deg')
    plt.semilogy(data[5, :,0], data[5, :,column] , "--", color = 'C2', label = '40 deg')


    plt.semilogy(data[6, :,0], data[6, :,column] , ":", color = 'C6', label = '0 deg')
    plt.semilogy(data[7, :,0], data[7,:,column]  , ":", color = 'C0', label = '20 deg')
    plt.semilogy(data[8, :,0], data[8,:,column] , ":", color = 'C2', label = '40 deg')

    labels =["0 deg", "20 deg", "40 deg"]

    h, l = ax.get_legend_handles_labels()
    ph = [plt.plot([],marker="", ls="")[0]]*3 # empty lines for 3 altititdes

    handles = ph[0:1] + h[0:3] + ph[1:2] + h[3:6] + ph[2:3] + h[6:9]
    labels = ["28'750km"] + l[0:3] + ["23'200km"] + l[3:6] + ["19'300km"] + l[6:9]
    leg = plt.legend(handles, labels, ncol=3)

    for vpack in leg._legend_handle_box.get_children():
        for hpack in vpack.get_children()[:1]:
            hpack.get_children()[0].set_width(0)

    if(xscale == 'log'):
        plt.xscale('log')



    plt.legend(handles, labels, ncol=3)
    #plt.ylim(0,1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.savefig(saving_name, dpi = 300)

def plot_ratio(data, xlabel = "x", ylabel = "y", title = None, saving_name = None, xscale = None, column = 1):
    fig, ax = plt.subplots()
    plt.plot(data[1, :,0], data[1, :,column] / data[0, :,column] , "-", color = 'C0', label = '20 deg')
    plt.plot(data[2, :,0], data[2, :,column] / data[0, :,column] , "-", color = 'C2', label = '40 deg')

    plt.plot(data[4, :,0], data[4, :,column] / data[3, :,column]  ,  "--", color = 'C0', label = '20 deg')
    plt.plot(data[5, :,0], data[5, :,column] / data[3, :,column] , "--", color = 'C2', label = '40 deg')

    plt.plot(data[7, :,0], data[7,:,column] / data[6, :,column]  , ":", color = 'C0', label = '20 deg')
    plt.plot(data[8, :,0], data[8,:,column] / data[6, :,column]  , ":", color = 'C2', label = '40 deg')

    labels =["20 deg", "40 deg"]

    h, l = ax.get_legend_handles_labels()
    ph = [plt.plot([],marker="", ls="")[0]]*3 # empty lines

    handles = ph[0:1] + h[0:2] + ph[1:2] + h[2:4] + ph[2:3] + h[4:6]
    labels = ["28'750km"] + l[0:2] + ["23'200km"] + l[2:4] + ["19'300km"] + l[4:6]
    leg = plt.legend(handles, labels, ncol=3)

    for vpack in leg._legend_handle_box.get_children():
        for hpack in vpack.get_children()[:1]:
            hpack.get_children()[0].set_width(0)

    if(xscale == 'log'):
        plt.xscale('log')



    plt.legend(handles, labels, ncol=3)
    plt.ylim(0,2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.savefig(saving_name, dpi = 300)
```
